```python
# Подключение по параметрам
import psycopg2
import json

def read_config():
    with open('conf.json', 'r') as file:
        config = json.load(file)
    return config

conf = read_config()
connection = psycopg2.connect(
    user = conf['user'],
    password = conf['password'],
    host = conf['host'],
    port = conf['port'],
    database = conf['database']
)
```
